```python
happiness = happiness_df.drop(columns = ['Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'])
print(happiness)
```

```python
happiness.rename(columns = {'Ladder score':'LadderScore', 'Country name' : 'CountryName', 'Regional indicator' : 'Region', 'Logged GDP per capita' : 'GDP', 'Social support' : 'Support', 'Healthy life expectancy' : 'LifeExpectancy', 'Freedom to make life choices' : 'Freedom', 'Perceptions of corruption' : 'Corruption'}, inplace = True) 
print(happiness)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(16, 8))  # Adjust the figsize as needed
fig.tight_layout()
fig.suptitle('Happiness Score', y=1.08)

axs[0, 0].plot(happiness['LadderScore'], happiness['GDP'])
axs[0, 0].set_title("GDP")

axs[0, 1].plot(happiness['LadderScore'], happiness['Support'])
axs[0, 1].set_title("Support")

axs[0, 2].plot(happiness['LadderScore'], happiness['LifeExpectancy'])
axs[0, 2].set_title("Life Expectancy")

axs[1, 0].scatter(happiness['LadderScore'], happiness['Freedom'])
axs[1, 0].set_title("Freedom")

axs[1, 1].plot(happiness['LadderScore'], happiness['Generosity'])
axs[1, 1].set_title("Generosity")

axs[1, 2].scatter(happiness['LadderScore'], happiness['Corruption'])
axs[1, 2].set_title('Corruption')

plt.show()
```
